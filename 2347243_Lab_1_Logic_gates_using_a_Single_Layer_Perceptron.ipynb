# AND Gate Classification
Scenario:
You are tasked with building a simple neural network to simulate an AND gate using a Single Layer Perceptron. The AND gate outputs 1 only if both inputs are 1; otherwise, it outputs 0.
Lab Task: Implement a Single Layer Perceptron using Python in Google Colab to classify the output of an AND gate given two binary inputs (0 or 1). Follow these steps:
Create a dataset representing the truth table of the AND gate.
Define the perceptron model with one neuron, including the activation function and weights initialization(Try both random weights and defined weights).
Train the perceptron using a suitable learning algorithm (e.g., gradient descent).
Test the model with all possible input combinations and display the results.

```python
import numpy as np

X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 0, 0, 1])

weights = np.random.rand(2)
bias = np.random.rand(1)
learning_rate = 0.1
epochs = 10

def step_function(x):
    return 1 if x >= 0 else 0

print(f'Initial weights: {weights}')
print(f'Initial bias: {bias}\n')

for epoch in range(epochs):
    print(f'Epoch {epoch+1}:')
    all_correct = True
    for i in range(len(X)):
        weighted_sum = np.dot(X[i], weights) + bias
        y_pred = step_function(weighted_sum)
        error = y[i] - y_pred

        if error != 0:
            all_correct = False
            weights += learning_rate * error * X[i]
            bias += learning_rate * error

        print(f'Input: {X[i]}, Predicted: {y_pred}, Actual: {y[i]}, Error: {error}')
        print(f'Updated Weights: {weights}, Updated Bias: {bias}\n')

    if all_correct:
        print('Training complete, all outputs are correct.')
        print(f'Final Weights: {weights}, Final Bias: {bias}')
        break
    print('-' * 30)

print('\nFinal Results after training:')
for i in range(len(X)):
    weighted_sum = np.dot(X[i], weights) + bias
    y_pred = step_function(weighted_sum)
    print(f'Input: {X[i]}, Predicted Output: {y_pred}')
```

# XOR Gate Classification
Scenario:
The XOR gate is known for its complexity, as it outputs 1 only when the inputs are different. This is a challenge for a Single Layer Perceptron since XOR is not linearly separable.
Lab Task: Attempt to implement a Single Layer Perceptron in Google Colab to classify the output of an XOR gate. Perform the following steps:
Create the XOR gate's truth table dataset.
Implement the perceptron model and train it using the XOR dataset.
Observe and discuss the perceptron's performance in this scenario.

```python
import numpy as np

X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 1, 1, 0])

weights = np.random.rand(2)
bias = np.random.rand(1)
learning_rate = 0.1
epochs = 10

def step_function(x):
    return 1 if x >= 0 else 0

print(f'Initial weights: {weights}')
print(f'Initial bias: {bias}\n')

for epoch in range(epochs):
    print(f'Epoch {epoch+1}:')
    all_correct = True
    for i in range(len(X)):
        weighted_sum = np.dot(X[i], weights) + bias
        y_pred = step_function(weighted_sum)
        error = y[i] - y_pred

        if error != 0:
            all_correct = False
            weights += learning_rate * error * X[i]
            bias += learning_rate * error

        print(f'Input: {X[i]}, Predicted: {y_pred}, Actual: {y[i]}, Error: {error}')
        print(f'Updated Weights: {weights}, Updated Bias: {bias}\n')

    if all_correct:
        print('Training complete, all outputs are correct.')
        print(f'Final Weights: {weights}, Final Bias: {bias}')
        break
    print('-' * 30)

print('\nFinal Results after training:')
for i in range(len(X)):
    weighted_sum = np.dot(X[i], weights) + bias
    y_pred = step_function(weighted_sum)
    print(f'Input: {X[i]}, Predicted Output: {y_pred}')
```

# Questions:
Why does the Single Layer Perceptron struggle to classify the XOR gate?

The XOR gate is not linearly separable. A Single Layer Perceptron can only create linear decision boundaries, which is insufficient for separating the inputs of the XOR gate into their respective outputs. XOR requires a more complex, nonlinear decision boundary.

What modifications can be made to the neural network model to handle the XOR gate correctly?

To handle XOR, we need to introduce non-linearity into the model. This can be achieved by using:

Multilayer Perceptron (MLP): A neural network with at least one hidden layer. This allows the network to learn more complex decision boundaries, making it capable of solving XOR.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from mlxtend.plotting import plot_decision_regions

X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 1, 1, 0])

mlp_random = MLPClassifier(hidden_layer_sizes=(2,),
                           activation='tanh',
                           max_iter=1,
                           solver='adam',
                           random_state=42)

mlp_random.fit(X, y)

mlp_trained = MLPClassifier(hidden_layer_sizes=(2,),
                            activation='tanh',
                            max_iter=10000,
                            solver='adam',
                            random_state=42,
                            learning_rate_init=0.01)

mlp_trained.fit(X, y)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

plot_decision_regions(X, y, clf=mlp_random, ax=ax1, legend=2)
ax1.set_title('Before Training')

plot_decision_regions(X, y, clf=mlp_trained, ax=ax2, legend=2)
ax2.set_title('After Training')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def step_function(x):
    return 1 if x >= 0 else 0

def plot_decision_boundary(weights, bias, gate_name, data, labels, epoch_label):
    x_values = np.linspace(-0.5, 1.5, 100)
    if weights[1] != 0:
        y_values = -(weights[0] / weights[1]) * x_values - (bias / weights[1])
    else:
        y_values = np.zeros_like(x_values)

    plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='bwr', edgecolor='k', s=100)
    plt.plot(x_values, y_values, 'k--', label="Decision Boundary")
    plt.title(f'{gate_name} Gate - {epoch_label}')
    plt.xlabel('Input 1')
    plt.ylabel('Input 2')
    plt.xlim(-0.5, 1.5)
    plt.ylim(-0.5, 1.5)
    plt.grid(True)
    plt.legend()
    plt.show()

gates_data = {
    "AND": {
        "inputs": np.array([[0, 0], [0, 1], [1, 0], [1, 1]]),
        "outputs": np.array([0, 0, 0, 1])
    },
    "OR": {
        "inputs": np.array([[0, 0], [0, 1], [1, 0], [1, 1]]),
        "outputs": np.array([0, 1, 1, 1])
    },
    "AND-NOT": {
        "inputs": np.array([[0, 0], [0, 1], [1, 0], [1, 1]]),
        "outputs": np.array([0, 0, 1, 0])
    },
    "XOR": {
        "inputs": np.array([[0, 0], [0, 1], [1, 0], [1, 1]]),
        "outputs": np.array([0, 1, 1, 0])
    }
}

random_weights = {
    "AND": np.random.rand(2),
    "OR": np.random.rand(2),
    "AND-NOT": np.random.rand(2),
    "XOR": np.random.rand(2)
}
random_bias = {
    "AND": np.random.rand(1),
    "OR": np.random.rand(1),
    "AND-NOT": np.random.rand(1),
    "XOR": np.random.rand(1)
}

trained_weights = {
    "AND": np.array([1, 1]),
    "OR": np.array([1, 1]),
    "AND-NOT": np.array([1, -1]),
    "XOR": np.array([1, 1])
}
trained_bias = {
    "AND": -1.5,
    "OR": -0.5,
    "AND-NOT": -0.5,
    "XOR": 0
}

for gate, data in gates_data.items():
    inputs = data['inputs']
    outputs = data['outputs']

    plot_decision_boundary(random_weights[gate], random_bias[gate], gate, inputs, outputs, 'Before Training')
    plot_decision_boundary(trained_weights[gate], trained_bias[gate], gate, inputs, outputs, 'After Training')
```
